```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8)
pd.options.mode.chained_assignment=None
```

```python
df=pd.read_csv('movies.csv')
```

```python
x=np.mean(df['genre'].isnull())
x
```

```python
#droping row with blank space
df.dropna(inplace=True)
```

```python
for col in df.columns:
    pc1_missing=np.mean(df[col].isnull())
    print('{}-{}%'.format(col,pc1_missing))
```

```python
df.dtypes
```

```python
#coverting object type to int 
df['budget']=df['budget'].astype('int64')
```

```python
df['gross']=df['gross'].astype('int64')
```

```python
df['votes']=df['votes'].astype('int64')
```

```python
df.sort_values('gross',ascending=False).head(5)
```

```python
pd.set_option('display.max_rows',None)
```

```python
#relation btw budget and gross
plt.scatter(x=df['budget'],y=df['gross'])
plt.title("budget vs gross")
plt.xlabel("budget")
plt.ylabel("gross")
plt.show()
```

```python
sns.regplot(x='budget',y='gross',data=df,scatter_kws={'color':'red'},line_kws={'color':'blue'})
```

```python
#correlation using pearson 
df.corr(method="pearson")#pearson,kendall,spearson
```

```python
#heatmap of correlation matrix
correlation_matrix=df.corr(method="pearson")
sns.heatmap(correlation_matrix,annot=True)
plt.title('correlation matrix')
plt.xlabel('movie features')
plt.show()
```

```python
#coverting object to category and cat codes
df_numerized=df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes
df_numerized.head(5)
```

```python
df.head(5)
```

```python
correlation_matrix=df_numerized.corr(method="pearson")
sns.heatmap(correlation_matrix,annot=True)
plt.title('correlation matrix')
plt.xlabel('movie features')
plt.ylabel('movie features')
plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat=df_numerized.corr()
corr_pairs=correlation_mat.unstack()
```

```python
corr_pairs.head(5)
```

```python
pd.set_option('display.max_rows',None)
```

```python
#sorting correlation
sorted_pairs=corr_pairs.sort_values(ascending=False)
sorted_pairs.head(5)
```

```python
#relation btw gross and votes
plt.scatter(x=df['gross'],y=df['votes'])
plt.ylabel('votes')
plt.xlabel('gross')
plt.title('gross vs votes')
plt.show()
```

```python
#reggression plot 
sns.regplot(x='votes',y='gross',data=df,scatter_kws={'color':'red'},line_kws={'color':'blue'})
```

```python
#highest correlation
high_corr=sorted_pairs[(sorted_pairs)>.5]
high_corr
```
